```python
import gradio as gr
from transformers import pipeline
from sentence_transformers import SentenceTransformer, util
import os
from dotenv import load_dotenv
```

```python
model='Babelscape/wikineural-multilingual-ner'
HF_TOKEN = os.getenv("HF_TOKEN")

ner = pipeline(
            "token-classification",
            model=model,
            aggregation_strategy="simple",
            token=HF_TOKEN,
        )

# Интерфейс с вкладками
def ner_text(text):
    entities = ner(text)
    list_ent=[]
    for ent in entities:
        list_ent.append(f"  {ent['word']:20s} → {ent['entity_group']:5s} ({ent['score']:.0%})")
    return list_ent

with gr.Blocks(title="AI Toolkit Demo") as demo:
    #gr.Markdown("# AI Toolkit: Извлечение сущностей")

   text_in = gr.Textbox(label="Текст", lines=2)
   text_out = gr.Textbox(label="Результат")
   gr.Button("Анализировать").click(ner_text, text_in, text_out)
demo.launch()

```
