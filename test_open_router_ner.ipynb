```python
import json
import time
import re
import os
from openai import OpenAI
from eval_tasks_ner import EVAL_TASKS_NER

from dotenv import load_dotenv

load_dotenv()
import json

OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
```

```python
model="openai/gpt-oss-120b:free"
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=OPENROUTER_API_KEY,
)
```

```python
results=[]
for entry in EVAL_TASKS_NER:
    article = entry['text']
    # Измерение времени
    start_time = time.perf_counter()
    prompt=f"Выдели сущности в тексте. Ответ в виде json.\n Текст:{article}"
    response = client.chat.completions.create(
                    model=model,
                    messages=[{"role": "user", "content":  prompt}],
                )
    elapsed_time = time.time() - start_time
    
    answer = response.choices[0].message.content
    print(answer)        
    results.append({
            "text": article,
            "ner_model": answer,
            "ner_test": entry['entities'],
            "elapsed_time": elapsed_time
            })


model_results_file=r'H:\netology\for_test_out\openai_gpt-oss-120b_free_results.json'
# Сохранение результатов в JSON файл для каждой модели
with open(model_results_file, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=4)
```
