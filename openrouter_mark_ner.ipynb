```python
from openai import OpenAI
import os
import time
from dotenv import load_dotenv

load_dotenv()
import json

OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
```

```python
system = """
Твоя задача: оценить качество извлечения сущностей из русского текста путем сравнения извлеченных сущностей с эталоном.
"""

instruction = """
Оцени качество выполнения задания по извлечению сущностей из русского текста по 10 бальной шкале где 10 отлично и 0 очень плохо. Форма ответа - это число от 0 до 10. В ответе должна быть только цифра.
"""
```

```python
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=OPENROUTER_API_KEY,
)


def test_model(ner_model,ner_test):
    messages = [
        {"role": "system", "content": system},
        {"role": "user", "content": f"{instruction}.\n\Эталон:\n{ner_test}\n\nСущности:\n"
                                        f"{ner_model}."}
    ]
    response = client.chat.completions.create(
                    model="openai/gpt-oss-120b:free",
                    messages=messages,
                )
    answer = response.choices[0].message.content
    print(answer)
    return answer
```

```python
path_res=r'H:\netology\1'
path_out=r'H:\netology\for_test_out'
if not os.path.exists(path_out):
    os.makedirs(path_out)

for filename in os.listdir(path_res):
    print(filename)
    file_path=os.path.join(path_res, filename)
    base_name = os.path.splitext(filename)[0]
    file_path_out=os.path.join(path_out, base_name+"_acc.json")

    with open(file_path, 'r', encoding='utf-8') as file:
        data=json.load(file)
        for record in data:        
            record['mark'] = test_model(record['ner_model'],record['ner_test'])
        with open(file_path_out, 'w', encoding='utf-8') as file_out:
            json.dump(data, file_out, ensure_ascii=False, indent=4)    
    
```

```python
import re

path_out=r'H:\netology\lesson5\out'
path_test_files=r'H:\netology\lesson5\for_test_out'
if not os.path.exists(path_out):
    os.makedirs(path_out)
results=[]
# Используем регулярное выражение для извлечения нужной част
for filename in os.listdir(path_test_files):
    print(filename)
    file_path=os.path.join(path_test_files, filename)
    base_name = os.path.splitext(filename)[0]
    match = re.search(r'ner_(\w+)_results_acc', base_name)
    if match:
        base_name = match.group(1)
    with open(file_path, 'r', encoding='utf-8') as file:
        data=json.load(file)
        mark_total=0
        time_total=0
        for record in data:        
            mark_total += int(record['mark'])
            time_total += float(record['elapsed_time'])
    results.append({
            "ai": base_name,
            "acc": mark_total,
            "time": time_total,
            })        
            
           
```

```python
print("\nРезультаты:")
for result in results:
    print(f"AI: {result['ai']}, Accuracy: {result['acc']}, Time: {result['time']}")
```

```python
print("\nРезультаты:")
print(f"{'AI':<50} {'Accuracy':<10} {'Time':<10}")
print("-" * 60)

for result in results:
    print(f"{result['ai']:<50} {result['acc']:<10} {result['time']:<10.2f}")
```
